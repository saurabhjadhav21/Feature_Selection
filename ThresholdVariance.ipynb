```python
from sklearn.feature_selection import VarianceThreshold
import pandas as pd
```

```python
data = pd.DataFrame({"A":[4,25,4,50],
                   "B":[0,0,0,0],
                   "C":[23,25,26,28],
                   "D":[0.0,0.0,0,0]})
```

```python
data
```

```python
var_thr=VarianceThreshold(threshold=0)
var_thr.fit(data)
```

```python
var_thr.get_support()
```

```python
data.columns[var_thr.get_support()]
```

```python
colums_remove =[colums for colums in data.columns
               if colums not in data.columns[var_thr.get_support()]]
```

```python
data.drop(colums_remove,axis=1)
```
